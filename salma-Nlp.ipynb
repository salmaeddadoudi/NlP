```python
import re
from collections import Counter
import numpy as np
import pandas as pd
```

```python
#importation et transformation du text en list de mots
def process_data():
    f = open('shakespeare.txt', 'r')
    content = f.read()
    f.close()    
    content=content.lower()
    #trensformer tous les mots du text et les stocker dans la liste list_word 
    list_word = re.findall('\w+',content)
   
    return list_word
```

```python
#en fait appelle a la fonction 
liste=process_data()
```

```python
#ajouter les mots unique du corpus dans vocab
vocab=set(liste)
print(len(vocab))
vocab
```

```python
#cette methode permet de cree un dict qui contien les mots comme cle est leur nombre d'occurence

def get_count(liste):
    word_count_dict ={} 
    for word in liste:
        if word in word_count_dict.keys():
             word_count_dict[word]=word_count_dict[word]+1
        else:
             word_count_dict[word]=1
   
    return word_count_dict
```

```python
word_count_dict=get_count(liste)
word_count_dict
```

```python
#cette foction reurn dictionnaire qui contien les mots comme cle et leur probabilter dans le corpus
def get_probs(word_count_dict):
    probs = {}  
    
    m = sum(word_count_dict.values())
 
    for word in word_count_dict.keys():
        probs[word] = word_count_dict[word] / m
    return probs
```

```python
probs=get_probs(word_count_dict)
probs
```

string manipulation

```python
#étant donné un mot, il renvoie toutes les chaînes possibles dont un caractère a été supprimé.
def delete_letter(word,verbose=False):
    delete_l = []
    split_l = []
    
    for c in range(len(word)):
        split_l.append((word[:c],word[c:]))
    for a,b in split_l:
        delete_l.append(a+b[1:])


    #if verbose: print(f"input word {word}, \nsplit_l = {split_l}, \ndelete_l = {delete_l}")

    return delete_l
```

```python
delete_word_l = delete_letter("cans",verbose=True)
```

```python
#renvoie une liste de tous les chaînes possibles de deux lettres adjacentes
def switch_letter(word, verbose=False):
 
    switch_l = []
    split_l = []
  
    len_word=len(word)
    for c in range(len_word):
        split_l.append((word[:c],word[c:]))
    switch_l = [a + b[1] + b[0] + b[2:] for a,b in split_l if len(b) >= 2]

    
    #if verbose: print(f"Input word = {word} \nsplit_l = {split_l} \nswitch_l = {switch_l}") 

    return switch_l
```

```python
switch_word_l = switch_letter(word="eta",verbose=True)
```

```python
#étant donné un mot, il renvoie toutes les chaînes possibles qui ont un caractère remplacé par une autre lettre différente
def replace_letter(word):

    letters = 'abcdefghijklmnopqrstuvwxyz'
    replace_l = []
    split_l = []
    
    for c in range(len(word)):
        split_l.append((word[0:c],word[c:]))
        
    replace_l = [a + l + (b[1:] if len(b)> 1 else '') for a,b in split_l if b for l in letters]
    replace_set=set(replace_l)    
    replace_set.remove(word)

    replace_l = sorted(list(replace_set))
    
    #print(f"Input word = {word} \nsplit_l = {split_l} \nreplace_l {replace_l}")   
    
    return replace_l
```

```python
replace_l = replace_letter(word='can')
```

```python
#donné un mot, il renvoie toutes les chaînes possibles qui ont un caractère supplémentaire inséré
def insert_letter(word):

    letters = 'abcdefghijklmnopqrstuvwxyz'
    insert_l = []
    split_l = []

    for c in range(len(word)+1):
        split_l.append((word[0:c],word[c:]))
    insert_l = [ a + l + b for a,b in split_l for l in letters]

    #print(f"Input word {word} \nsplit_l = {split_l} \ninsert_l = {insert_l}")
    
    return insert_l
```

```python
#prend en entrée un mot et retourne un ensemble de tous les mots qui peuvent être obtenus en effectuant une seule modification
def edit_one_letter(word, allow_switches = True):

    edit_one_set = set()

    edit_one_set.update(delete_letter(word))
    if allow_switches:
        edit_one_set.update(switch_letter(word))
    edit_one_set.update(replace_letter(word))
    edit_one_set.update(insert_letter(word))

    return edit_one_set
```

```python
tmp_word = "hello"
tmp_edit_one_set = edit_one_letter(tmp_word)
# turn this into a list to sort it, in order to view it
tmp_edit_one_l = sorted(list(tmp_edit_one_set))

print(f"input word {tmp_word} \nedit_one_l \n{tmp_edit_one_l[:10]}\n")
print(f"The type of the returned object should be a set {type(tmp_edit_one_set)}")
print(f"Number of outputs from edit_one_letter('at') is {len(edit_one_letter('at'))}")
```

```python
#prend en entrée un mot et retourne un ensemble de tous les mots qui peuvent être obtenus en effectuant deux modification
def edit_two_letters(word, allow_switches = True):

    
    edit_two_set = set()
    edit_one = edit_one_letter(word,allow_switches=allow_switches)
    for w in edit_one:
        if w:
            edit_two = edit_one_letter(w,allow_switches=allow_switches)
            edit_two_set.update(edit_two)

    return edit_two_set
```

```python
tmp_edit_two_set = edit_two_letters("hello")
tmp_edit_two_l = sorted(list(tmp_edit_two_set))
print(f"Number of strings with edit distance of two: {len(tmp_edit_two_l)}")
print(f"First 10 strings {tmp_edit_two_l[:10]}")
print(f"Last 10 strings {tmp_edit_two_l[-10:]}")
print(f"The data type of the returned object should be a set {type(tmp_edit_two_set)}")
```

```python
#retourne une liste des n meilleures corrections possibles pour le mot donné, triées par ordre décroissant de probabilité
def get_corrections(word, probs, vocab, n=2, verbose = False):

    
    suggestions = []
    n_best = []
    
    suggestions = list((word in vocab and word) or edit_one_letter(word).intersection(vocab) or edit_two_letters(word).intersection(vocab))
    n_best = [[s,probs[s]] for s in list(reversed(suggestions))]

    
    if verbose: print("suggestions = ", suggestions)

    return n_best
```

```python
my_word = 'heven' 
tmp_corrections = get_corrections(my_word, probs, vocab, 2, verbose=True)

for i, word_prob in enumerate(tmp_corrections):
    print(f"word {i}: {word_prob[0]}, probability {word_prob[1]:.6f}")

print(f"data type of corrections {type(tmp_corrections)}")
```
